```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np

# Look pretty...

# matplotlib.style.use('ggplot')
plt.style.use('ggplot')
```

```python
df=pd.read_csv('crack-growth.csv', sep=',',header=0)
#df
```

```python
sl=df.loc[0:,['Delta K','Frequency','da/dN']]
sl
```

```python
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection ="3d")

# plot should be done on the `axis`: ax
x=df["Delta K"]
y=df["Frequency"]
z=df["da/dN"]
ax.scatter3D(x,y,z,marker='.',c='r')
# Set labels for axex and title for plot
#ax.plot3D(x,y,z,marker='',c='g',label='Linear Regression Model')

ax.set_xlabel('Delta K')
ax.set_ylabel('Frequency')
ax.set_zlabel('da/dN')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots() 

fig.set_figwidth(15)
fig.set_figheight(15)
# plot should be done on the `axis`: ax
x=df["Delta K"]
z=df["da/dN"]
ax.scatter(x, z,marker='.',c='r',label='Entire Data')
# Set labels for axex and title for plot

ax.set_xlabel('Delta K')
ax.set_ylabel('da/dN')
ax.set_title('Delta K vs da/dN')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots() 

fig.set_figwidth(15)
fig.set_figheight(15)
# plot should be done on the `axis`: ax
y=df["Frequency"]
z=df["da/dN"]
ax.scatter(x, z,marker='.',c='b',label='Entire Data')
# Set labels for axex and title for plot

ax.set_xlabel('Frequency')
ax.set_ylabel('da/dN')
ax.set_title('Frequency vs da/dN')
ax.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

from sklearn.model_selection import train_test_split

x_values=sl[["Delta K","Frequency"]]
#X
y_values=sl["da/dN"]
#Y

X_train, X_test, y_train, y_test = train_test_split(x_values,y_values, test_size= 0.30)

Polynomial_Regression_model = Pipeline([('poly', PolynomialFeatures(degree = 14)),('linear', LinearRegression(fit_intercept=False))])
Polynomial_Regression_model.fit(X_train,y_train)

score_train = Polynomial_Regression_model.score(X_train, y_train)
score_test = Polynomial_Regression_model.score(X_test, y_test)
print('R^2 score for training data = %10.4f'%(score_train))
print('R^2 score for testing data  = %10.4f'%(score_test))
```
